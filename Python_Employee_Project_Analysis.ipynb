```python
import pandas as pd  # Import pandas for data manipulation
import numpy as np  # Import numpy for numerical operations
```

```python
project = pd.DataFrame({
    'ID': ['A001','A002','A003','A004','A005','A002','A005','A003','A001','A003','A001','A004','A004','A005'],
    'Project': ['Project 1','Project 2','Project 3','Project 4','Project 5','Project 6','Project 7','Project 8','Project 9','Project 10','Project 11','Project 12','Project 13','Project 14'],
    'Cost': [1002000,2000000,4500000,5500000,None,680000,400000,350000,None,300000,2000000,1000000,3000000,200000],
    'Status': ['Finished','Ongoing','Finished','Ongoing','Finished','Failed','Finished','Failed','Ongoing','Finished','Failed','Ongoing','Finished','Finished']
}) # Create Project dataframe from given data
```

```python
project
```

```python
employee = pd.DataFrame({
    'ID': ['A001','A002','A003','A004','A005'],
    'Name': ['John Alter','Alice Luxumberg','Tom Sabestine','Nina Adgra','Amy Johny'],
    'Gender': ['M','F','M','F','F'],
    'City': ['Paris','London','Berlin','Newyork','Madrid'],
    'Age': [25,27,29,31,30]
}) # Create employee dataframe from given data
```

```python
employee
```

```python
seniority = pd.DataFrame({
    'ID': ['A001','A002','A003','A004','A005'],
    'Designation Level': [2,2,3,2,3]
}) # Create seniority dataframe from given data
```

```python
seniority
```

```python
project.to_csv('project.csv', index=False)
employee.to_csv('employee.csv', index=False)
seniority.to_csv('seniority.csv', index=False) # Save project, employee, seniority data to CSV file
```

```python
for i in range(len(project)):
    if pd.isna(project.loc[i, 'Cost']):
        project.loc[i, 'Cost'] = project['Cost'][:i].mean()
```

```python
project
```

```python
employee[['First Name','Last Name']] = employee['Name'].str.split(' ', expand=True) # Split full name into first and last name columns
```

```python
employee.drop('Name', axis=1, inplace=True) # Remove original Name column after splitting
```

```python
final = pd.merge(project, employee, on='ID') # Merge project and employee data using ID
```

```python
final = pd.merge(final, seniority, on='ID') # Merge previous result with seniority data using ID
```

```python
final
```

```python
final['Bonus'] = np.where(final['Status'] == 'Finished',
                          final['Cost'] * 0.05,
                          0)  # Give 5% bonus if project is finished, else 0
```

```python
final
```

```python
final.loc[final['Status'] == 'Failed', 'Designation Level'] -= 1 # Decrease designation level by 1
```

```python
final = final[final['Designation Level'] <= 4]
```

```python
final
```

```python
final['First Name'] = np.where(final['Gender'] == 'M',
                              'Mr. ' + final['First Name'],
                              'Mrs. ' + final['First Name']) # Assign Mr. for males and Mrs. for females
```

```python
final.drop('Gender', axis=1, inplace=True)
```

```python
final
```

```python
final.loc[final['Age'] > 29, 'Designation Level'] += 1 # Increase designation level by 1 if age is more than 29
```

```python
final
```

```python
total_cost = final.groupby(['ID','First Name'])['Cost'].sum().reset_index() # Calculate total cost of projects for each employee ID
```

```python
total_cost.rename(columns={'Cost':'Total Cost'}, inplace=True)
```

```python
total_cost
```
